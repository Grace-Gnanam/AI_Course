```python
import pandas as pd
df = pd.read_csv('insurance_pre.csv')
df.head(6)
```

```python
print("Rows:",df.shape[0],"\nColumns:",df.shape[1])
```

```python
# Counting un-missing data
df.isna().count()
```

```python
df.dtypes
```

```python
print("Unique values in each column")
print("Sex:",df['sex'].unique())
print("Children:",df['children'].unique())
print("Smoker:",df['smoker'].unique())
```

```python
print("Verifying max and min values in numerical columns")
print("Age:",df['age'].min(),"to",df['age'].max())
print("BMI:",df['bmi'].min(),"to",df['bmi'].max())
print("Charges:",df['charges'].min(),"to",df['charges'].max())
```

```python
from sklearn.preprocessing import LabelEncoder
le_1 = LabelEncoder()
df['sex'] = le_1.fit_transform(df['sex'])
le_2 = LabelEncoder()
df['smoker'] = le_2.fit_transform(df['smoker'])

df.head()
```

```python
# Saving Cleaned data
df.to_csv('cleaned_data.csv', index=False)
```
