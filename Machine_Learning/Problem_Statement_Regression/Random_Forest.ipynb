```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
df = pd.read_csv('cleaned_data.csv')
```

```python
independent = df[['age','sex','bmi','children','smoker']]
dependent = df[['charges']]
```

```python
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=1/3,random_state=43)


regressor=RandomForestRegressor(n_estimators=100,criterion='poisson',max_features='sqrt')
regressor=regressor.fit(X_train,y_train)

y_pred=regressor.predict(X_test)

r_score=r2_score(y_test,y_pred)
print("R2 Score:",r_score)
```

```python
import pickle
filename="Random_Forest.sav"
pickle.dump(regressor,open(filename,"wb"))
```
