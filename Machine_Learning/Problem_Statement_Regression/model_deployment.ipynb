```python
import pickle
model=pickle.load(open("Random_Forest.sav","rb"))
```

```python
age_input=int(input("Enter your age:"))
gender_input=int(input("Enter your Gender number 1.Female 2.Male:"))
bmi_input=float(input("Enter your BMI index:"))
children_input = int(input("Enter number of children :"))
smoker_input = int(input("Are you a smoker 1.No 2.Yes:"))
                         
result=model.predict([[age_input,gender_input-1,bmi_input,children_input,smoker_input-1]])
print("The Charges predicted is ₹",round(result[0],2))
```
